## Popularity based recommendation system

```python
#we will show the top 50 books having highest average rating along with atleast 250 rating count
```

```python
# *Merging of Datasets*

rating_with_name =rating.merge(books,on="ISBN")
```

```python
rating_with_name.head()
```

```python
rating_with_name.shape
```

```python
num_rating=rating_with_name.groupby("Book-Title")["Book-Rating"].count().reset_index()
num_rating.rename(columns={"Book-Rating":"num Rating"},inplace= True)
num_rating
```

```python
avg_rating=rating_with_name.groupby("Book-Title")["Book-Rating"].mean().reset_index()
avg_rating.rename(columns={"Book-Rating":"Avg Rating"},inplace= True)
avg_rating
```

```python
popular_df = num_rating.merge(avg_rating,on="Book-Title")
popular_df
```

```python
top50=popular_df[popular_df["num Rating"]>250].sort_values("Avg Rating",ascending=False).head(50).reset_index(drop=True)
```

```python
top50=top50.merge(books,on="Book-Title").drop_duplicates("Book-Title")[["Book-Title","Book-Author","Image-URL-M","num Rating","Avg Rating"]]
```

```python
top50=top50.reset_index(drop=True)
top50
```

## collaborative based recommender system

```python
##only consider experienced readers that is they have rated atleast 200 books and also considering those books having minimum 50 books
```

```python
rating_with_name
```

```python
readers=rating_with_name.groupby("User-ID")["Book-Rating"].count().reset_index()
readers[readers["Book-Rating"]>250]
readers=readers["User-ID"]
readers
```

```python
# y=filtered_rating.groupby("Book-Title").count()["Book-Rating"]>=50
# famous_books=y[y].index
# # famous_books=y[y["Book-Rating"]>50]["Book-Title"]
# famous_books

r=rating_with_name.groupby("User-ID").count()["Book-Rating"]>250
readers=r[r].index
readers
```

```python
filtered_rating=rating_with_name[rating_with_name["User-ID"].isin(readers)]
filtered_rating
```

```python
y=filtered_rating.groupby("Book-Title").count()["Book-Rating"]>=50
famous_books=y[y].index
# famous_books=y[y["Book-Rating"]>50]["Book-Title"]
famous_books
```

```python
final_rating=filtered_rating[filtered_rating["Book-Title"].isin(famous_books)].copy()
final_rating.drop_duplicates(inplace=True)
```

```python
temp=final_rating[["User-ID","Book-Title","Book-Rating"]].copy()
temp
```

```python
temp.drop_duplicates(inplace=True)
```

```python
pt=temp.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
# temp.nunique()
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
# famous_books.shape, readers.shape, filtered_books.shape, filtered_rating.shape,filtered_rating.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
similarity_score.shape
```

```python
similarity_score[0]
```

```python
# so this similarity score is containing all the similarity of columns with another available column
# now creating function that will give us the top 5 similarity score on speicific given book name
```

```python
def detail(book_name):
    temp=books[books["Book-Title"]==book_name]
    item=[]
    item.append(list(temp["Book-Title"])[0])
    item.append(list(temp["Book-Author"])[0])
    item.append(list(temp["Image-URL-M"])[0])
    return item

def recommend(book_name):
    index=np.where([pt.index==book_name])[1][0]
    distance=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)
    similar_books=distance[1:6]
    book_list=[pt.index[x[0]] for x in similar_books]
    book_list=list(map(lambda x:detail(x),book_list))
    return book_list
```
